## Using Multiple Linear Regression to Predict CO2 Emissions

```python
X = co2_scaled
Y = co2['CO2 (mg C m¯² d¯¹)']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

### Baseline Model

```python
from sklearn.dummy import DummyRegressor

dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(x_train, y_train)
dummy_regr.predict(x_train)
baseline = dummy_regr.score(x_train, y_train)
print("Baseline R^2: %f" %baseline)
```

### Multiple Linear Regression - 1st fit

```python
from sklearn import linear_model
ols = linear_model.LinearRegression()

ols.fit(x_train,y_train)
print("Coeficients:")
print(pd.DataFrame(zip(x_train.columns, ols.coef_)).sort_values(by=1))
print("R^2: %f" %(ols.score(x_train, y_train)))
```

### Before moving on, lets check the assumptions of Multiple Linear Regression:
    1. No Multicolinearity
    2. Normality of Residuals
    3. Constant Variance
    4. Independent Errors
    5. Linear relationships


```python
#Multicolinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
pd.DataFrame({'vif': vif[0:]}, index=x_train.columns).T
```

In order to adhere to no multicolinearity, we must DROP seasonal features

```python
x_train = x_train.drop(['NDVI_spring_avg_lc','NDVI_summer_avg_lc','NDVI_fall_avg_lc',
                       'temp_spring_avg_lc', 'temp_fall_avg_lc', 'temp_summer_avg_lc', 'temp_winter_avg_lc',
                       'npp_spring_avg_lc', 'npp_summer_avg_lc', 'npp_winter_avg_lc', 'npp_fall_avg_lc'], axis=1)

x_test = x_test.drop(['NDVI_spring_avg_lc','NDVI_summer_avg_lc','NDVI_fall_avg_lc',
                       'temp_spring_avg_lc', 'temp_fall_avg_lc', 'temp_summer_avg_lc', 'temp_winter_avg_lc',
                       'npp_spring_avg_lc', 'npp_summer_avg_lc', 'npp_winter_avg_lc', 'npp_fall_avg_lc'], axis=1)

vif = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
pd.DataFrame({'vif': vif[0:]}, index=x_train.columns).T
```

```python
#Normality of Residual
import seaborn as sns

ols.fit(x_train,y_train)
residual = y_train - ols.predict(x_train)

sns.distplot(residual)
```

```python
#Constant Variance / Homoscedasticity

plt.scatter(ols.predict(x_train), residual)

#errors are too high compared w prediction (near 100% for some)
```

### Penalized Regression using Ridge

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.set_params(alpha=.01, normalize=False)
ridge.fit(x_train,y_train)
print('The intercept is %.4f' %(ridge.intercept_))
ridgeCoef = pd.Series(ridge.coef_, index=x_train.columns)
print('The slopes are \n%s' %(ridgeCoef.sort_values()))
ridge.score(x_train, y_train)
```

```python
ridge.score(x_test, y_test)
```

```python
ridge_coefs = []
ridge_intercepts = []
R2  = []
alphaRange = np.linspace(1e-10,10,25)
for alpha in alphaRange:
    ridge.set_params(alpha=alpha, normalize = False)  
    ridge.fit(x_train, y_train)
    ridge_intercepts.append(ridge.intercept_)
    ridge_coefs.append(ridge.coef_)
    R2.append(ridge.score(x_train, y_train))
```

```python
ridge_coefs_matrix = pd.DataFrame(ridge_coefs, columns=x_train.columns)
ridge_coefs_matrix['alpha'] = alphaRange
ridge_coefs_matrix['R2'] = R2
ridge_coefs_matrix
```
