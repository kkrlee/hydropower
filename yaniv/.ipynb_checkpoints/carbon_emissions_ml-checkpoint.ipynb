## Pre-Processing Data for Machine Learning

```python
for column in co2:
    print(column, ':',  co2[column].isna().sum())
```

```python
co2['kWh'].hist()
co2['kWh'].describe()
```

```python
co2['Volume'].hist()
co2['Volume'].describe()
```

```python
from sklearn.impute import KNNImputer
model_impute = KNNImputer(n_neighbors=int(np.sqrt(co2.shape[0])))
co2_imputed = model_impute.fit_transform(co2)
```

```python
co2_imputed = pd.DataFrame(columns=co2.columns, data=co2_imputed)
```

```python
#check
for column in co2_imputed:
    print(column, ':',  co2_imputed[column].isna().sum())
```

```python
co2['Volume'].hist(color='orange', alpha=.5)
co2_imputed['Volume'].hist(color='blue', alpha=.5)
```

```python
co2['kWh'].hist(color='orange', alpha=.5)
co2_imputed['kWh'].hist(color='blue', alpha=.5)
```

### standardize data

```python
from sklearn.preprocessing import StandardScaler

data = co2_imputed.drop(['CO2 (mg C m¯² d¯¹)'], axis=1).copy()

scaler = StandardScaler()

co2_scaled = pd.DataFrame(scaler.fit_transform(data),columns=co2.columns[1:])
```

```python
co2_scaled.head()
```

## Using Multiple Linear Regression to Predict CO2 Emissions

```python
X = co2_scaled
Y = co2['CO2 (mg C m¯² d¯¹)']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

### Baseline Model

```python
from sklearn.dummy import DummyRegressor

dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(x_train, y_train)
dummy_regr.predict(x_train)
baseline = dummy_regr.score(x_train, y_train)
print("Baseline R^2: %f" %baseline)
```

```python
from sklearn import linear_model
ols = linear_model.LinearRegression()

ols.fit(x_train,y_train)
print("Intercept: %f" %ols.intercept_)
print("Coefficients: %s" %str(ols.coef_))
print("R^2: %f" %(ols.score(x_train, y_train)))
```

```python
import statsmodels.api as sm

model = sm.OLS(y_train, x_train)
results = model.fit()
print(results.summary())
```

### Tuning the model

```python
##remove Tmin and Tmax
X = X.drop(['Tmin','Tmax'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

ols2 = linear_model.LinearRegression()

ols2.fit(x_train,y_train)
print("Intercept: %f" %ols2.intercept_)
print("Coefficients: %s" %str(ols2.coef_))
print("R^2: %f" %(ols2.score(x_train, y_train)))
```

```python
model = sm.OLS(y_train, x_train)
results = model.fit()
print(results.summary())
```

### Penalized Regression using Ridge

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.set_params(alpha=1, normalize=False)
ridge.fit(x_train,y_train)
print('The intercept is %.4f' %(ridge.intercept_))
ridgeCoef = pd.Series(ridge.coef_, index=x_train.columns)
print('The slopes are \n%s' %(ridgeCoef.sort_values()))
ridge.score(x_train, y_train)
```
